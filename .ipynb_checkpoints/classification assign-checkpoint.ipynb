```python
import pandas as pd
import numpy as np

import sklearn
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Diabetes.csv')
```

```python
df
```

```python
sns.jointplot(x='age', y='bloodpressure', data=df)
```

```python
sns.jointplot(x='age', y='insulin', data=df)
```

```python
sns.scatterplot(x='age', y='insulin',hue='outcome', data=df)
```

```python
Target = df['outcome']
Target 
```

```python
target_counts = df['outcome'].value_counts()
```

```python
target_counts.values
```

```python
sns.barplot(x=target_counts.index, y=target_counts.values)
plt.xlabel('Target Class')
plt.ylabel('Count')
```

```python
Dataset = df.drop(columns=['outcome'])
Dataset
```

```python
X_train, X_test, y_train, y_test = train_test_split(Dataset, Target, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train,y_train)
accuracy = knn.score(X_test, y_test)
```

```python
print("Accuracy: ", accuracy)
```

```python
K_pred = knn.predict(X_test)
knn_cm = confusion_matrix(y_test, K_pred)
```

```python
sns.heatmap(knn_cm, annot=True, fmt='d')
plt.title('Confusion Matrix knn')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
k_range = list(range(1, 31))
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train,y_train, cv=10, scoring='accuracy')
    k_scores.append(scores.mean())
```

```python
mean_knn, std_knn = np.mean(k_scores), np.std(k_scores)
```

```python
mean_knn
```

```python
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
```

```python
best_k = k_range[np.argmax(k_scores)]
best_k
```

```python
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
```

```python
accuracy = clf.score(X_test, y_test)
print("Accuracy: ", accuracy)
```

```python
y_pred = clf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True, fmt='d')
plt.title('Confusion Matrix svc')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
scores_svc = cross_val_score(clf,X_test, y_test, cv=5)
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores_svc.mean(), scores_svc.std() * 2))
```

```python
plt.plot(np.arange(len(scores_svc)), scores_svc , 'o-')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-Validation Scores')
plt.show()
```

```python
scores_svc_tain = cross_val_score(clf,X_train, y_train, cv=5)
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores_svc_tain.mean(), scores_svc_tain.std() * 2))
```

```python
plt.plot(np.arange(len(scores_svc_tain)), scores_svc_tain , 'o-')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-Validation Scores')
plt.show()
```

```python
log = LogisticRegression(penalty='l1', solver='saga', max_iter=10000)
log.fit(X_train, y_train)
```

```python
l_pred = log.predict(X_test)
```

```python
accuracy_L = log.score(X_test, y_test)
print("Accuracy:", accuracy_L)
```

```python
cm_L = confusion_matrix(y_test, l_pred)
```

```python
sns.heatmap(cm_L, annot=True, fmt='d')
plt.title('Confusion Matrix lr')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
scores_lr_train = cross_val_score(log,X_train, y_train, cv=5)
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores_lr_train.mean(), scores_lr_train.std() * 2))
```

```python
scores_lr_train.std()
```

```python
scores_lr_train.mean()
```

```python
plt.plot(np.arange(len(scores_lr_train)), scores_lr_train , 'o-')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.title('Cross-Validation Scores')
plt.show()
```

```python
bar_width = 0.35

plt.bar(0,scores_svc_tain.mean(), bar_width, yerr=scores_svc_tain.std(), color='r', label='SVC')
plt.bar(bar_width,scores_lr_train.mean() , bar_width, yerr=scores_lr_train.std(), color='g', label='Logistic Regression')
plt.bar(bar_width*2,mean_knn , bar_width, yerr=std_knn, color='b', label='KNN')

plt.xlabel('Classifier')
plt.ylabel('Mean Accuracy Score (+/- Standard Deviation)')
plt.title('Cross-Validation Accuracy Scores')

plt.xticks([0, 0.35,0.70], ['SVC', 'Logistic Regression','KNN'])
plt.legend(loc='lower right')

plt.show()
```
